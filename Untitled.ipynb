```python
def yearly_genre_popularity(data, year, genre):
    """
    Plot genre popularity and display total votes for a given year and genre.
    
    Parameters:
    data (DataFrame): The dataset containing columns 'startYear', 'genres', and 'numVotes'.
    year (int or str): The year to filter by.
    genre (str): The genre category to filter by.
    """
    # Filter data for the specified year and genre
    filtered_data = data[(data['startYear'] == int(year)) & (data['genres'].str.contains(genre, case=False))]
    
    # Calculate total votes for the specified genre in the given year
    total_votes = filtered_data['numVotes'].sum()
    
    # Plot the data
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=filtered_data, x='startYear', y='numVotes', hue='genres', legend=False)
    plt.title(f'Genre Popularity in {year} for {genre} (Total Votes: {total_votes})')
    plt.xlabel('Year')
    plt.ylabel('Number of Votes')
    
    # Display total votes on the plot
    plt.text(int(year), max(filtered_data['numVotes'], default=0), 
             f'Total Votes: {total_votes}', fontsize=12, ha='center', va='bottom')
    
    plt.show()
yearly_genre_popularity(split_genre, '2014', 'Drama')
```
